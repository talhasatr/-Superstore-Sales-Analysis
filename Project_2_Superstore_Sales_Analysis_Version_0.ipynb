# Feature Engineering

```python
data.info()
```

```python
data['shipping_timeframe'] = data['Ship Date'] - data['Order Date']
data[['shipping_timeframe', '_']]= data['shipping_timeframe'].astype('str').str.split(expand=True)
data['shipping_timeframe'] = data['shipping_timeframe'].astype(int)
```

```python
data[data['shipping_timeframe']<0]
```

```python
# It doesn't make sense to have ship date earlier than the order date. This is a data mistake and we need to fix it
data['Ship Mode'].value_counts()
```

```python
data['Ship Mode'][data['shipping_timeframe']<0].value_counts()
```

```python
data[data.shipping_timeframe>=0].groupby(['Ship Mode'])['shipping_timeframe'].agg(['median'])
```

```python
# standard class 6 days
# Same day: 0 days
# First class: 3 days
# Second class: 4 days
```

```python
# let's create a function to fix this:
def ship_date_fix(ship_mode, add_day):
  data['Ship Date'] = np.where((data.shipping_timeframe < 0) & (data['Ship Mode'] == ship_mode),
                               data['Order Date'] + timedelta(days=add_day),
                               data['Ship Date'])
  
```

```python
# let's write a for loop take all ship modes and corresponding days:
ship_mode_dict = {'Standard Class': 6,
                  'Second Class': 4,
                  'First Class': 3,
                  'Same Day':0}
for key, value in ship_mode_dict.items():
  ship_date_fix(key, value)
```

```python
data['shipping_timeframe'] = data['Ship Date'] - data['Order Date']
data[['shipping_timeframe', '_']]= data['shipping_timeframe'].astype('str').str.split(expand=True)
data['shipping_timeframe'] = data['shipping_timeframe'].astype(int)
```
